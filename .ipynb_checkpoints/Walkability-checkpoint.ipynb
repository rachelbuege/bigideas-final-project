```python
import pandas as pd
import matplotlib.pyplot as plt

walk = pd.read_csv("https://data.wprdc.org/dataset/4d3d4324-b32a-4519-b56f-cee2340057bf/resource/682b1df1-a63b-4413-9362-ba077af63baa/download/walkscorect.xls-walk-score-by-ct.csv")

sortedwalk = walk.sort_values(by='Walk Score', ascending=False)

plt.style.use('seaborn')
walkplot = sortedwalk.head(10).plot(fontsize = '14',
                                    kind='bar', 
                                    y='Walk Score',
                                    x='Census Tract',
                                    rot = 0,
                                    ylabel='Walk Score', 
                                    xlabel='Census Tract', 
                                    title='Top 10 Census Tracts by Walk Score', 
                                    figsize=(10,10))
```
